```python
import pandas as pd

gen = pd.read_csv("Plant_1_Generation_Data.csv")
weather = pd.read_csv("Plant_1_Weather_Sensor_Data.csv")

gen['DATE_TIME'] = pd.to_datetime(gen['DATE_TIME'], errors='coerce')
weather['DATE_TIME'] = pd.to_datetime(weather['DATE_TIME'], errors='coerce')

gen['DATE_TIME'] = gen['DATE_TIME'].dt.floor('15min')
weather['DATE_TIME'] = weather['DATE_TIME'].dt.floor('15min')

df = pd.merge(gen, weather, on="DATE_TIME")

df['is_fault'] = ((df['IRRADIATION'] > 0.05) & (df['DC_POWER'] < 20)).astype(int)

df.columns = df.columns.str.strip()   # IMPORTANT FIX

df.to_csv("cleaned_solar_data.csv", index=False)

print(df.columns)
print("CSV created successfully")
print(df.shape)
```
